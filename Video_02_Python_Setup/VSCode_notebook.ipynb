Wir erstellen einen Plot einer Sinus- und einer Cosinus Kurve

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 2.0 * np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label=r'$\sin(x)$')
plt.plot(x, y2, label=r'$\cos(x)$')
plt.xlabel(r'$x$')
plt.ylabel(r'f(x)')
plt.legend()
```
