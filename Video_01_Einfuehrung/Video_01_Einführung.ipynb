# Python für (angehende) Naturwissenschaftler

## Vorkenntnisse:

- Grundkenntnisse in der Programmierung mit Python
- Kenntnisse in der Verwendung von Jupyter Notebooks / JupyterLab

Am besten haben Sie schon mathematische/wissenschaftliche Probleme gelöst und **eigene** Projekte verwirklicht!

[Project Euler](https://projecteuler.net/) ist eine hervorragende Quelle, wenn Sie mathematische Probleme suchen. Die ersten zehn Probleme sollten Sie lösen können.

Vorschlag für Materialien zum Selbststudium (Programmieranfänger, Grundkenntnisse in anderen Programmiersprachen): [Think Python von Allen Downey](https://allendowney.github.io/ThinkPython/index.html)  (3. Auflage; Kapitel 1-13)

## Python-Werkzeuge für wissenschaftliches Programmieren

- Python-Entwicklungsumgebung: [Python Distribution Anaconda](https://www.anaconda.com/) und [Codeeditor VSCode](https://code.visualstudio.com/)
- Das [NumPy Modul](https://numpy.org/) (**sehr effiziente** numerische Datenstrukturen: Vektoren, Matrizen, multidimensionale Arrays)
- Das [SciPy Modul](https://scipy.org/) (Umfangreiche Sammlung numerischer Algorithmen auf Basis von NumPy, u. a. für Integration, Lösung von Differentialgleichungen, Optimierung, Fits und lineare Algebra)
- Das [Matplotlib Modul](https://matplotlib.org/) (2D und 3D Visualisierung wissenschaftlicher Daten)
- Das [SymPy Modul](https://www.sympy.org/en/index.html) (Symbolische Mathematik in Python, z. B. für Ableitungen, Integrale, Vereinfachungen und symbolische Lösungen von Gleichungen)
- Versionskontrolle mit [Git](https://git-scm.com/) und [GitHub](https://github.com/) (Versionsverwaltung wissenschaftlicher Projekte, Nachverfolgung von Codeänderungen, Organisation und Sicherung vom Programmcode)

**Suchen Sie sich eigene Projekte und setzen Sie diese in Python um!**
